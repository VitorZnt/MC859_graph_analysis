```python
import os
import openai

client = openai.OpenAI(api_key=os.environ.get("MARITACA_API_KEY"),
                       base_url="https://chat.maritaca.ai/api")
```

```python
prompt_system = """Você é um classificador de desinformação. Analise o texto fornecido e classifique o assunto principal na melhor categoria. Sua resposta deve conter o número de uma das categorias, somente o número.

1. Política e Eleições: Fraude em urnas, ataque a instituições públicas.
2. Economia: Impostos, inflação, crise, PIX, Bolsa Família, financiamento.
3. Saúde e Ciência: Vacinas, tratamentos, alimentação, crise sanitária, pseudociências, ceticismo sobre ciência.
4. Segurança e Golpes: Link de prêmio, alarmismo sobre crimes e golpes.
5. Sociedade: Preconceitos, discriminação de minorias, pessoas e etnias.
6. Calamidades: Invasão, apocalipse, guerra.
7. Indeterminado: apenas se não houver relação com algum assunto anterior.
"""

# test cases
prompt_user=[]

# chat 37445 - expected classification 3
prompt_user.append("""As propriedades do alecrim são maravilhosas. Desinflama, desinfeta e cicatriza. Cozinhar alguns galhos de alecrim com agua morna e mergulhar a patinha do seu animalzinho, lavar a ferida, deixá-la na água por cerca de 10 minutos, você verá melhora no ferimento. Não importa se seu bichinho lambe para secar a patinha, alecrim tranquiliza, é bom para digestão e eliminar parasitas Vídeo de Joana Darc De Lima Pés no chão, Cabeça erguida  E fé na vida.  Domingou🌼 A lava se move a 700m / h na zona do Parque Industrial, às 13:30 (hora das Ilhas Canárias). NÃO IMPORTA!  Não importa o quão belo(a) você é. Todos temos o mesmo fim. É só questão de tempo. Portanto antes de se sentir superior aos outros por causa da aparência, pense duas vezes. O fato é que não importa se tu é rico ou pobre, analfabeto ou formado em faculdade. Todos viraremos pó. A única diferença é: PARA ONDE VOCÊ VAI APÓS MORRER. O 'hotspot' subaquático que alimenta o vulcão de La Palma criará novas ilhas  O fluxo de magma que moldou o arquipélago das Canárias espanholas há 20 milhões de anos continua a adicionar massa de terra, enquanto Fuerteventura e Lanzarote estão destinadas a afundar sob os efeitos da erosão Vídeo de Marilene Bezerra de Medei""")

# chat 40269 - expected classification 5
prompt_user.append("""🇺🇸🏳‍⚧🏳‍🌈 Uma escola de ensino médio da Pensilvânia instalou mictórios no banheiro feminino.  👉 @amidianaoesuaamiga 🧠""")

# chat 40371 - expected classification 6
prompt_user.append("""A Turquia disse a aliados que rejeitará a adesão da Suécia e da Finlândia à Otan, afirmou o presidente turco, Tayyip Erdogan, em um vídeo postado em sua conta no Twitter nesta quinta-feira (19).""")
```

```python
def get_response(n):
    response = client.chat.completions.create(
        model="sabia-4",
        messages=[
            {"role": "system", "content": prompt_system},
            {"role": "user", "content": prompt_user[n]}
        ],
        stream=False,
        tools=[],
        max_tokens=5,
        temperature=0.0
    )
    return response
```
