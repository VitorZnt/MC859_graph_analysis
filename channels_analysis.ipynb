```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
def freeman_centrality(df):
    n = len(df)
    c_max = df['ch.incomingDegree'].iloc[0]

    sum_degree_dif = (c_max - df['ch.incomingDegree']).sum()
    
    divisor = (n-1) * (n-2)
    
    return sum_degree_dif / divisor
```

```python
df = pd.read_csv("analysis_data/channels_graph/channels_incoming_degree.csv") # this csv is ordered desc
centr = freeman_centrality(df)
print(centr)
```
